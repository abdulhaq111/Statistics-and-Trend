```python
#Calculating Statistical Analysis Of Dataset for Each country
grouped_df = df.groupby('Country Name')

# Loop through each group (country)
for country, group in grouped_df:
    print(f"Country: {country}")
    
    # Loop through each series name
    for series_name in df['Series Name'].unique():
        # Filter the group for the current series name
        series_data = group[group['Series Name'] == series_name].iloc[:, 4:]
        
        # Perform statistical operations
        if not series_data.empty:
            print(f"\nSeries Name: {series_name}")
            print("Mean:", series_data.mean().mean())
            print("Median:", series_data.median().median())
            print("Standard Deviation:", series_data.std().std())
            print("Describe:")
            print(series_data.describe())
            print("\nCorrelation Matrix:")
            print(series_data.corr())
            print("\n")
```

```python
#Evaluating Skewness
grouped_df = df.groupby('Country Name')

# Iterate over unique country names
for country_name, country_data in grouped_df:
    # Filter the data for the current country and GDP growth series
    gdp_growth_country = country_data[(country_data['Series Name'] == 'GDP growth (annual %)')]
    
    # If there's no data for GDP growth, continue to the next country
    if gdp_growth_country.empty:
        continue
    
    # Extract the years and corresponding GDP growth values
    years = [int(col.split(' ')[0]) for col in gdp_growth_country.columns[4:]]
    gdp_growth_values = gdp_growth_country.iloc[0, 4:].astype(float)
    skewness = gdp_growth_values.skew()
    
    print(f"Skewness of GDP growth (annual %) for {country_name}: {skewness}")
```

```python
#Evaluating Kurtosis
grouped_df = df.groupby('Country Name')

# Iterate over unique country names
for country_name, country_data in grouped_df:
    # Filter the data for the current country and GDP growth series
    gdp_growth_country = country_data[(country_data['Series Name'] == 'GDP growth (annual %)')]
    
    # If there's no data for GDP growth, continue to the next country
    if gdp_growth_country.empty:
        continue
    
    # Extract the years and corresponding GDP growth values
    years = [int(col.split(' ')[0]) for col in gdp_growth_country.columns[4:]]
    gdp_growth_values = gdp_growth_country.iloc[0, 4:].astype(float)
    
    # Calculate the kurtosis of GDP growth data for the current country
    kurtosis = gdp_growth_values.kurt()
    
    print(f"Kurtosis of GDP growth (annual %) for {country_name}: {kurtosis}")
```

# Catagorical Graph

```python
# 1) Catagorical Graph to Visualize GDP of Countries

def plot_gdp_growth(countries):
    """
    Plots the GDP growth for selected countries over the years 2018-2022.

    Parameters:
    - countries (list): A list of country names for which GDP growth is to be plotted.

    Returns:
    - None
    """
    # Select years for comparison
    years = ['2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]', '2022 [YR2022]']

    # Defing custom color palette
    colors = ['steelblue', 'mediumseagreen', 'firebrick', 'gold', 'slategray', 'darkorange']

    # Filter the DataFrame to include only selected countries and years
    filtered_df = df[df['Country Name'].isin(countries) & df['Series Name'].str.contains('GDP growth')]

    # Pivot the DataFrame for plotting
    pivot_df = filtered_df.pivot(index='Country Name', columns='Series Name', values=years)

    # Convert the data to numeric values
    pivot_df = pivot_df.apply(pd.to_numeric, errors='coerce')

    # Plot the grouped bar chart
    ax = pivot_df.plot(kind='bar', figsize=(12, 8), edgecolor='black', color=colors)
    plt.title('GDP Growth for Selected Countries (2018-2022)', fontsize=16)
    plt.xlabel('Country', fontsize=14)
    plt.ylabel('GDP Growth (annual %)', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(years, title='Year', fontsize=12, loc='lower left')
    plt.grid(False)  # Remove grid

    # Add horizontal line at x=0
    plt.axhline(y=0, color='black', linestyle='-')

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
# 2) Catagorical Graph to Visualize Inflaton Rate of Countries

def plot_inflation(df, countries):
    """
    Plots the inflation for selected countries over the years 2018-2022.

    Parameters:
    - df (DataFrame): The DataFrame containing the economic data.
    - countries (list): A list of country names for which inflation is to be plotted.

    Returns:
    - None
    """
    # Select years for comparison
    years = ['2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]', '2022 [YR2022]']

    # Define custom color palette
    colors = ['indigo', 'mediumseagreen', 'purple', 'darksalmon', 'firebrick']
    
    # Filter the DataFrame to include only selected countries and years
    filtered_df = df[df['Country Name'].isin(countries) & df['Series Name'].str.contains('Inflation, consumer prices')]

    # Pivot the DataFrame for plotting
    pivot_df = filtered_df.pivot(index='Country Name', columns='Series Name', values=years)

    # Convert the data to numeric values
    pivot_df = pivot_df.apply(pd.to_numeric, errors='coerce')

    # Plot the grouped bar chart
    ax = pivot_df.plot(kind='bar', figsize=(12, 8), edgecolor='black', color=colors)
    plt.title('Inflation for Selected Countries (2018-2022)', fontsize=16)
    plt.xlabel('Country', fontsize=14)
    plt.ylabel('Inflation (annual %)', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(years, title='Year', fontsize=12, loc='upper left')  # Adjusted legend position
    plt.grid(False)  # Remove grid

    # Add horizontal line at y=0
    plt.axhline(y=0, color='black', linestyle='-')

    # Show the plot
    plt.tight_layout()
    plt.show()
```

# Relational Graph

```python
# 1) Line Graph To Compare Total employers over years in various countries

def plot_employer_total(df, countries, figsize=(12, 8)):
    """
    Plots the comparison of employer's total (% of total employment) between multiple countries for the years 2018-2022.

    Parameters:
    - df (DataFrame): The DataFrame containing the data.
    - countries (list): A list of country names for which employer's total (% of total employment) is to be plotted.
    - figsize (tuple, optional): The size of the figure (width, height). Default is (12, 8).

    Returns:
    - None
    """
    # Define colors for the countries
    colors = ['b', 'r', 'g', 'c', 'm', 'y']
    
    plt.figure(figsize=figsize)  # Set the figure size
    
    for i, country in enumerate(countries):
        # Filter the DataFrame for the current country and select the employer's total (% of total employment) series
        country_employer_data = df[(df['Country Name'] == country) & 
                                   (df['Series Name'] == 'Employers, total (% of total employment) (modeled ILO estimate)')]

        # Extract the years and corresponding data for the current country
        years = country_employer_data.columns[4:]  # Assuming the years start from column index 4
        data = country_employer_data.iloc[0, 4:].astype(float)

        # Plot the line graph for the current country with dotted lines and no markers
        plt.plot(years, data, color=colors[i], linestyle='--', label=country)

    # Add title and labels
    plt.title("Employer's Total (% of Total Employment) Comparison between Multiple Countries (2018-2022)")
    plt.xlabel("Year")
    plt.ylabel("Employer's Total (% of Total Employment)")
    plt.grid(True)
    plt.xticks(rotation=45)

    # Move the legend outside the plot
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    plt.tight_layout()
    plt.show()
```

```python
# 1) Line Graph To Compare Total Internet users over years in various countries

def plot_internet_usage(df, countries, figsize=(12, 8)):
    """
    Plots the comparison of individuals using the internet (% of population) between multiple countries for the years 2018-2022.

    Parameters:
    - df (DataFrame): The DataFrame containing the data.
    - countries (list): A list of country names for which internet usage is to be plotted.
    - figsize (tuple, optional): The size of the figure (width, height). Default is (12, 8).

    Returns:
    - None
    """
    # Define colors for the countries
    colors = ['b', 'r', 'g', 'c', 'm', 'y']
    
    plt.figure(figsize=figsize)  # Set the figure size
    
    for i, country in enumerate(countries):
        # Filter the DataFrame for the current country and select the Individuals using the Internet series
        country_internet_data = df[(df['Country Name'] == country) & 
                                   (df['Series Name'] == 'Individuals using the Internet (% of population)')]

        # Extract the years and corresponding data for the current country
        years = country_internet_data.columns[4:]  # Assuming the years start from column index 4
        data = country_internet_data.iloc[0, 4:].astype(float)

        # Plot the line graph for the current country with dotted lines
        plt.plot(years, data, color=colors[i], linestyle='--', label=country)

    # Add title and labels
    plt.title("Individuals using the Internet (% of Population) Comparison between Multiple Countries (2018-2022)")
    plt.xlabel("Year")
    plt.ylabel("Individuals using the Internet (% of Population)")
    plt.grid(True)
    plt.xticks(rotation=45)

    # Move the legend outside the plot
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    plt.tight_layout()
    plt.show()
```

```python
def shorten_series_name(series_name):
    """
    Shortens the given series name for better readability.

    Parameters:
    - series_name (str): The full name of the series.

    Returns:
    - str: The shortened version of the series name.
    """
    # Define a dictionary to map full names to shorter names
    short_names_map = {
        'GDP growth (annual %)': 'GDP %',
        'Literacy rate, adult total (% of people ages 15 and above)': 'Literacy %',
        'Trade (% of GDP)': 'Trade % of GDP',
        'Central government debt, total (% of GDP)': 'Govt. Debt % of GDP',
        'Foreign direct investment, net (BoP, current US$)': 'FDI Net (US$)',
        'Individuals using the Internet (% of population)': 'Internet Users %',
        'Net trade in goods (BoP, current US$)': 'Net Trade (US$)',
        'Agriculture, forestry, and fishing, value added (% of GDP)': 'Agri GDP %',
        'Inflation, consumer prices (annual %)': 'Inflation %',
        'Access to electricity (% of population)': 'Access to Electricity %',
        'Employers, total (% of total employment) (modeled ILO estimate)': 'Employers %'
    }
    return short_names_map.get(series_name, series_name)

def plot_country_correlation(df: pd.DataFrame, country_name: str):
    """
    Plots a correlation heatmap for the given country using economic data.

    Parameters:
    - df (pd.DataFrame): DataFrame containing the economic data.
    - country_name (str): The name of the country for which correlation is to be analyzed.

    Returns:
    - None
    """
    # Filter rows for the specified country
    country_data = df[df['Country Name'] == country_name]

    # Define the years
    years = ['2017', '2018', '2019', '2020', '2021', '2022']

    # Create a list of country names
    country_names = [country_name] * country_data.shape[0] * len(years)

    # Create a list of unique series names
    series_names = country_data['Series Name'].unique()

    # Fetch values of series names for each year
    values = []
    for year in years:
        for series_name in series_names:
            col_name = f'{year} [YR{year}]'
            series_values = country_data.loc[country_data['Series Name'] == series_name, col_name].values
            series_values_cleaned = [float(value) if str(value) != ".." else 0 for value in series_values]
            values.extend(series_values_cleaned)

    # Create DataFrame
    data = {
        'Country Name': country_names,
        'Series Name': [name for name in series_names for _ in range(len(years))],
        'Year': [year for _ in range(len(series_names)) for year in years],
        'Value': values
    }

    df = pd.DataFrame(data)

    # Pivot DataFrame
    df_pivot = df.pivot(index=['Country Name', 'Year'], columns='Series Name', values='Value')

    # Compute correlation matrix
    corr_matrix = df_pivot.corr()

    # Plot heatmap
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, cmap='BuPu', fmt=".2f")

    # Shorten the series names for better readability on x-axis
    plt.xticks(ticks=range(len(series_names)), labels=[shorten_series_name(name) for name in series_names], rotation=45, ha='right')

    # Shorten the series names for better readability on y-axis
    plt.yticks(ticks=range(len(series_names)), labels=[shorten_series_name(name) for name in series_names], rotation=0, ha='right')

    plt.title(f'Correlation Heatmap for {country_name} (Years: 2017-2022)')
    plt.show()
```
